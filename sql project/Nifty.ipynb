```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
Nifty=pd.read_csv('Nifty_50.csv')
```

```python
# converting date datatype (object-datetime64)
Nifty['Date']=pd.to_datetime(Nifty['Date'])
```

```python
# changing col names and order
Nifty.rename(columns={'Price':'Closing Price','Open':'Opening Price'},inplace=True)
Nifty=Nifty.reindex(['Date','Opening Price','Closing Price','Change_in_percent','pe','pb','div_yield_percent'],axis=1)
Nifty
```

```python
# Insight 1
# Biggest daily gain (Top 10)
Nifty.set_index('Date').loc[:,['Change_in_percent']]\
                       .sort_values('Change_in_percent',ascending=False).head(10)
```

```python
# Insight 2
# Biggest daily losses (Top 10)
Nifty.set_index('Date').loc[:,['Change_in_percent']]\
                       .sort_values('Change_in_percent',ascending=True).head(10)
```

```python
# Insight 3
# Biggest intraday gains (Top 10)
Nifty_i3=Nifty.assign(Intraday_change=Nifty['Closing Price']-Nifty['Opening Price'])
Nifty_i3=Nifty_i3.assign(percent_change=(Nifty_i3['Intraday_change']/Nifty_i3['Opening Price']*100).round(2))
Nifty_i3.set_index('Date').loc[:,['Opening Price','Closing Price','percent_change']]\
                          .sort_values('percent_change',ascending=False).head(10)
```

```python
# Insight 4
# Biggest intraday losses (Top 10)
Nifty_i3=Nifty.assign(Intraday_change=Nifty['Closing Price']-Nifty['Opening Price'])
Nifty_i3=Nifty_i3.assign(percent_change=(Nifty_i3['Intraday_change']/Nifty_i3['Opening Price']*100).round(2))
Nifty_i3.set_index('Date').loc[:,['Opening Price','Closing Price','percent_change']]\
                          .sort_values('percent_change',ascending=True).head(10)
```

```python
# Insight 5 
# Average yearly returns and yearly returns ranked by years
Nifty_i5=Nifty.assign(Year=Nifty['Date'].dt.year)
Nifty_i5=Nifty_i5.groupby(['Year']).agg(Opening=('Closing Price','first'),
                                        Closing=('Closing Price','last'))
Nifty_i5=Nifty_i5.assign(Change=Nifty_i5['Opening']-Nifty_i5['Closing'])
Nifty_i5=Nifty_i5.assign(Change_percent=(Nifty_i5['Change']/Nifty_i5['Opening']*100).round(2))
# mean calculation
mean=Nifty_i5['Change_percent'].mean().round(2)
mean_2=Nifty_i5.loc[~Nifty_i5.index.isin([2020,2021]),'Change_percent'].mean().round(2)
# printing of mean
print(f'Average yearly returns: {mean} percent\nAverage yearly returns after removing outliers (Covid years): {mean_2}')
# Yearly Returns
Nifty_i5[['Opening','Closing','Change_percent']].sort_values('Change_percent',ascending=False)
```

```python
# Insight 6  (Method -1 - without multi-indexing)
# Highest Monthly returns (Top 10)
Nifty_i6=Nifty.assign(Month=Nifty['Date'].dt.month,
                      Month_name=Nifty['Date'].dt.month_name(),
                      Year=Nifty['Date'].dt.year
                      )
Nifty_i6=Nifty_i6.groupby(['Year','Month','Month_name'],as_index=False).agg(Opening=('Closing Price','first'),
                                                                            Closing=('Closing Price','last'))
Nifty_i6=Nifty_i6.assign(Change=Nifty_i6['Opening']-Nifty_i6['Closing'])
Nifty_i6=Nifty_i6.assign(Monthly_Returns=(Nifty_i6['Change']/Nifty_i6['Opening']*100).round(2))
Nifty_i6.loc[:,['Year','Month','Month_name','Monthly_Returns']].sort_values('Monthly_Returns',ascending=False).head(10)
```

```python
# Insight 6  (Method -2 - with multi-indexing)
# Highest Monthly returns (Top 10)
Nifty_i6_1=Nifty.assign(Month=Nifty['Date'].dt.month,
                        Month_name=Nifty['Date'].dt.month_name(),
                        Year=Nifty['Date'].dt.year
                      )
Nifty_i6_1=Nifty_i6_1.groupby(['Year','Month','Month_name']).agg(Opening=('Closing Price','first'),
                                                                 Closing=('Closing Price','last'))
Nifty_i6_1=Nifty_i6_1.assign(Change=Nifty_i6_1['Opening']-Nifty_i6_1['Closing'])
Nifty_i6_1=Nifty_i6_1.assign(Monthly_Returns=(Nifty_i6_1['Change']/Nifty_i6_1['Opening']*100).round(2))
Nifty_i6_1[['Monthly_Returns']].sort_values('Monthly_Returns',ascending=False).head(10)
```

```python
# Insight 7
# Best months according to avg returns
Nifty_i7=Nifty.assign(Month=Nifty['Date'].dt.month,
                      Month_name=Nifty['Date'].dt.month_name(),
                      Year=Nifty['Date'].dt.year
                      )
Nifty_i7=Nifty_i7.groupby(['Year','Month','Month_name'],as_index=False).agg(Opening=('Closing Price','first'),
                                                                            Closing=('Closing Price','last'))
Nifty_i7=Nifty_i7.assign(Change=Nifty_i7['Opening']-Nifty_i7['Closing'])
Nifty_i7=Nifty_i7.assign(Monthly_Returns=(Nifty_i7['Change']/Nifty_i7['Opening']*100).round(2))
Nifty_i7.groupby(['Month_name']).agg(avg_returns=('Monthly_Returns','mean')).round(2).sort_values('avg_returns',ascending=False)
```

```python
# Insight 8
# Average P/E, P/B , Dividend yield
Pe_mean=Nifty['pe'].mean().round(2)
Pb_mean=Nifty['pb'].mean().round(2)
Div_yield_mean=Nifty['div_yield_percent'].mean().round(2)
print(f'Average P/E ratio is: {Pe_mean}\n'\
      f'Average P/B ratio is: {Pb_mean}\n'\
      f'Median dividend yield is: {Div_yield_mean}')
```

```python
#9 Insight 9
# Average P/E, P/B , Dividend yield (excluding 2020 and 2021)
Pe_mean=Nifty[~Nifty['Date'].dt.year.isin([2020,2021])].loc[:,'pe'].mean().round(2)
Pb_mean=Nifty[~Nifty['Date'].dt.year.isin([2020,2021])].loc[:,'pb'].mean().round(2)
Div_yield_mean=Nifty[~Nifty['Date'].dt.year.isin([2020,2021])].loc[:,'div_yield_percent'].mean().round(2)
print(f'Average P/E ratio is: {Pe_mean}\n'\
      f'Average P/B ratio is: {Pb_mean}\n'\
      f'Median dividend yield is: {Div_yield_mean}')
# Outliers were not extreme and barely affected the quality and reliability of dataset
```

```python
# Parameter 2 
# Date Range Analysis

try:
    a=input('Starting Date (YYYY-MM-DD): ')
    b=input('Ending Date (YYYY-MM-DD): ')
    a = pd.to_datetime(a) 
    b=pd.to_datetime(b) 
    Filtered_data=Nifty[Nifty['Date'].between(a,b)].sort_index(ascending=False)
    answer=pd.DataFrame()
    if a>=b:
        print('ERROR!!!  Ending data cannot be smaller or same as the starting date')
    else:
        answer=Filtered_data
        opening=answer['Closing Price'].head(1).values [0]
        closing=answer['Closing20 Price'].tail(1).values [0]
        returns=((closing-opening)/opening*100).round(2)
        average_pe=answer['pe'].mean().round(2)
        average_pb=answer['pb'].mean().round(2)
        average_div_yield=answer['div_yield_percent'].mean().round(2)
        print('\nBasic Description:\n')
        print(f'Closing price on first date: {opening}\n'\
              f'Closing price on last date: {closing}\n'\
              f'Returns in percentage: {returns}\n'\
              f'Average PE ratio: {average_pe}\n'\
              f'Average PB ratio: {average_pb}\n'\
              f'Average Dividend yield ratio: {average_div_yield}\n'
              f'total days market was open: {len(answer)}')
    display(Filtered_data)
except Exception:
       print("Invalid input. Make sure dates are in the correct format (YYYY-MM-DD) and are valid calendar dates.")
```

```python
# Parameter 3
# Year-Based Summary

try:
    Year=int(input('Type the year: '))
    Filtered_data=Nifty[Nifty['Date'].dt.year==Year].sort_index(ascending=False)
    answer=pd.DataFrame()
    if (Year>2024) | (Year<2014):
        print('ERROR!!!  Select years from 2014-2024 only')
    else:
        answer=Filtered_data
        opening=answer['Closing Price'].head(1).values [0]
        closing=answer['Closing Price'].tail(1).values [0]
        returns=((closing-opening)/opening*100).round(2)
        average_pe=answer['pe'].mean().round(2)
        average_pb=answer['pb'].mean().round(2)
        average_div_yield=answer['div_yield_percent'].mean().round(2)
        print('\nBasic Description:\n')
        print(f'Closing price on first date: {opening}\n'\
              f'Closing price on last date: {closing}\n'\
              f'Returns in percentage: {returns}\n'\
              f'Average PE ratio: {average_pe}\n'\
              f'Average PB ratio: {average_pb}\n'\
              f'Average Dividend yield ratio: {average_div_yield}\n'
              f'total days market was open: {len(answer)}')
    display(Filtered_data)
except Exception:
       print("Invalid input. Make sure dates are in the correct format (YYYY-MM-DD) and are valid calendar dates.")
```
